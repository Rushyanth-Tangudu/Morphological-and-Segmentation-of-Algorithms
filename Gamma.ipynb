```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform power-law (gamma) transformation
def power_law_transformation(image, gamma, c=1):
    # Normalize image to the range [0, 1] for processing
    normalized_image = image / 255.0
    
    # Apply the power-law transformation formula
    gamma_corrected = c * np.power(normalized_image, gamma)
    
    # Normalize back to the range [0, 255] and convert to uint8
    gamma_corrected = np.uint8(255 * gamma_corrected)
    
    return gamma_corrected

# Load the grayscale image
image = cv2.imread(r"C:\Users\rushy\OneDrive\Desktop\AMRITA\Sem_5\DIP\dip_review_1\intensity\town-1509505.jpg", cv2.IMREAD_GRAYSCALE)

# List of gamma values to apply
gamma_values = [0.5, 1.0, 2.0]

# Set up the plot
plt.figure(figsize=(20, 5))

# Display the original image
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Perform gamma correction for each gamma value and display results
for i, gamma in enumerate(gamma_values):
    # Apply gamma correction
    gamma_image = power_law_transformation(image, gamma, c=1)
    
    # Display the result
    plt.subplot(1, 4, i + 2)
    plt.imshow(gamma_image, cmap='gray')
    plt.title(f'Gamma = {gamma}')
    plt.axis('off')
    
    # Save the result with a unique filename
    output_path = f'gamma_corrected_image_{gamma}.jpg'
    cv2.imwrite(output_path, gamma_image)
    print(f"Gamma-corrected image saved at: {output_path}")

plt.show()
```
