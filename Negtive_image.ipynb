```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform negative transformation
def negative_transformation(image, L=255):
    # Apply the negative transformation formula
    new_image = L - 1 - image
    
    # Return the transformed image
    return new_image

# Load the grayscale image
image = cv2.imread(r"C:\Users\rushy\OneDrive\Desktop\AMRITA\Sem_5\DIP\dip_review_1\intensity\OIP.jpeg", cv2.IMREAD_GRAYSCALE)

# Perform negative transformation
neg_image = negative_transformation(image)

# Display the original and negative images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(neg_image, cmap='gray')
plt.title('Negative Image')
plt.axis('off')

plt.show()

# Save the result
cv2.imwrite('negative_image.jpg', neg_image)
```
