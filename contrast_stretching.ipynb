```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform contrast stretching
def contrast_stretching(image, a, b):
    # Apply the contrast stretching formula
    new_image = a * image + b
    
    # Clip the values to keep them in the range [0, 255]
    new_image = np.clip(new_image, 0, 255)
    
    # Convert to uint8 (since image data is typically 8-bit)
    return new_image.astype(np.uint8)

# Load the grayscale image
image = cv2.imread(r"C:\Users\rushy\OneDrive\Desktop\AMRITA\Sem_5\DIP\dip_review_1\intensity\th.jpeg", cv2.IMREAD_GRAYSCALE)

# Parameters for contrast stretching
a = 1.5  # Scaling factor for contrast
b = 20   # Shifting factor for brightness

# Perform contrast stretching
stretched_image = contrast_stretching(image, a, b)

# Display the original and contrast-stretched images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(stretched_image, cmap='gray')
plt.title('Contrast-Stretched Image')
plt.axis('off')

plt.show()

# Save the result
cv2.imwrite('contrast_stretched_image.jpg', stretched_image)
```
