```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform bit-plane slicing
def bit_plane_slicing(image):
    # Initialize a list to hold the bit planes
    bit_planes = []
    
    # Extract each bit plane from the image
    for i in range(8):  # For 8-bit images
        # Shift right by i bits and isolate the least significant bit
        bit_plane = (image >> i) & 1
        bit_planes.append(bit_plane * 255)  # Scale to 255 for visualization

    return bit_planes

# Load the grayscale image
image = cv2.imread(r"C:\Users\rushy\OneDrive\Desktop\AMRITA\Sem_5\DIP\dip_review_1\intensity\OIP (11).jpeg", cv2.IMREAD_GRAYSCALE)

# Perform bit-plane slicing
bit_planes = bit_plane_slicing(image)

# Display the original image and bit planes
plt.figure(figsize=(15, 10))

# Show the original image
plt.subplot(3, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Display each bit plane
for i in range(8):
    plt.subplot(3, 3, i + 2)
    plt.imshow(bit_planes[i], cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')

plt.tight_layout()
plt.show()

# Optionally, save the bit planes as images
for i, bit_plane in enumerate(bit_planes):
    cv2.imwrite(f'bit_plane_{i}.jpg', bit_plane)
    print(f"Bit Plane {i} saved as 'bit_plane_{i}.jpg'")
```
