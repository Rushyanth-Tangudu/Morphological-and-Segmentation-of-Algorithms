```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform logarithmic transformation
def logarithmic_transformation(image, c=1):
    # Apply the logarithmic transformation formula
    log_image = c * np.log1p(image)  # np.log1p(x) is equivalent to log(1 + x)
    
    # Normalize the output to fit in the range [0, 255]
    log_image = np.uint8(255 * (log_image / np.max(log_image)))
    
    return log_image

# Load the grayscale image
image = cv2.imread(r"C:\Users\rushy\OneDrive\Desktop\AMRITA\Sem_5\DIP\dip_review_1\intensity\OIP (3).jpeg", cv2.IMREAD_GRAYSCALE)

# Perform logarithmic transformation
c = 1  # Constant for scaling (you can experiment with different values)
log_image = logarithmic_transformation(image, c)

# Display the original and log-transformed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(log_image, cmap='gray')
plt.title('Logarithmic Transformed Image')
plt.axis('off')

plt.show()

# Save the result
cv2.imwrite('logarithmic_image.jpg', log_image)
```
