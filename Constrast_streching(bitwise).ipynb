```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function for contrast stretching
def contrast_stretching(image, low_in, high_in, low_out, high_out):
    # Create a copy of the image to avoid modifying the original
    stretched_image = np.zeros_like(image)

    # Apply contrast stretching formula
    # Normalize and stretch intensity values
    mask = (image >= low_in) & (image <= high_in)
    stretched_image[mask] = ((image[mask] - low_in) * (high_out - low_out) / (high_in - low_in)) + low_out

    # For values below low_in, set to low_out
    stretched_image[image < low_in] = low_out
    # For values above high_in, set to high_out
    stretched_image[image > high_in] = high_out

    return np.uint8(stretched_image)

# Load the grayscale image
image = cv2.imread(r"C:\Users\rushy\OneDrive\Desktop\AMRITA\Sem_5\DIP\dip_review_1\intensity\OIP (1).jpeg", cv2.IMREAD_GRAYSCALE)

# Define the input and output intensity ranges
low_in = 50   # Low input intensity
high_in = 200  # High input intensity
low_out = 0    # Low output intensity
high_out = 255 # High output intensity

# Perform contrast stretching
stretched_image = contrast_stretching(image, low_in, high_in, low_out, high_out)

# Display the original and stretched images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(stretched_image, cmap='gray')
plt.title('Contrast Stretched Image')
plt.axis('off')

plt.show()

# Save the result
cv2.imwrite('contrast_stretched_image.jpg', stretched_image)
print("Contrast stretched image saved as 'contrast_stretched_image.jpg'")
```
