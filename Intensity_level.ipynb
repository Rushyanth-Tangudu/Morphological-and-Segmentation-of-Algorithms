```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function for intensity level slicing
def intensity_level_slicing(image, low, high, slice_value):
    # Create a copy of the original image
    sliced_image = np.zeros_like(image)
    
    # Apply intensity level slicing
    sliced_image[(image >= low) & (image <= high)] = slice_value

    return sliced_image

# Load the grayscale image
image = cv2.imread(r"C:\Users\rushy\OneDrive\Desktop\AMRITA\Sem_5\DIP\dip_review_1\intensity\OIP (9).jpeg", cv2.IMREAD_GRAYSCALE)

# Define the intensity range to highlight
low = 100   # Low intensity value to highlight
high = 150  # High intensity value to highlight
slice_value = 255  # Value to assign in the highlighted range

# Perform intensity level slicing
sliced_image = intensity_level_slicing(image, low, high, slice_value)

# Display the original and sliced images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sliced_image, cmap='gray')
plt.title('Intensity Level Sliced Image')
plt.axis('off')

plt.show()

# Save the result
cv2.imwrite('intensity_level_sliced_image.jpg', sliced_image)
print("Intensity level sliced image saved as 'intensity_level_sliced_image.jpg'")
```
